```python
import random
import numpy as np
random.seed(10)
we=np.random.random(1)
random.seed(10)
print(random.random())
print(we)
```

```python
# for various weights, we are calculating the expected return and volatility for the coreresponding portfolio.
por_ret1=[] 
por_vol1=[]
w1=[]
sr1=[]
count=0
a1= len(returns_comp1.columns)
for x in range(50000):
    weights1=[]
    for j in range(10):
        random.seed(count)
        weights1.append(random.random())
        count+=1
    weights1=np.array(weights1)
    weights1 /= np.sum(weights1)
    w1.append(weights1)
    por_ret1.append(np.sum(weights1*returns_comp1.mean())*250)
    por_vol1.append(np.sqrt(np.dot(weights1.T,np.dot(returns_comp1.cov()*250,weights1))))
       

port_r1=np.array(por_ret1)
port_vol1=np.array(por_vol1)
port_w1=np.array(w1)
port_r1,port_vol1,port_w1
for i in range(50000):
    sr1.append((port_r1[i]-0.06)/port_vol1[i])
```

```python
# Creating a new dataframe for weights
arr2=np.array(range(10))
df1=pd.DataFrame(port_w1,columns=[arr2])
```

```python
portfolio1=pd.DataFrame({"Return" :port_r1, "Volatility":port_vol1,"Sr":sr1})
portfolio1
```

```python
#Finding out the maximum sharpe ratio
max(portfolio1['Sr'])
```

```python
# Concating the weight dataframe with existing one
df_new1 = pd.concat([portfolio1, df1], axis=1, join='inner')
df_new1
```
